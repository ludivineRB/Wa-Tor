```python
#importation
import csv
import matplotlib as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#RÃ©cupÃ©ration du fichier 
data = pd.read_csv('data_WaTor.csv', sep=',')
data
```

```python
data.columns
data.rename(index=str, columns={"nb_fish":"nombre de poisson", "nb_shark":"nombre de requin"}, inplace=True)
data
data["nombre total animaux"]=data["nombre de poisson"]+data["nombre de requin"]
data
```

```python
plt.figure(figsize=(20,10)) 
plt.title("Number of Sharks and Fish")
plt.plot(data["chronon"], data["nombre de poisson"], color="lightblue", label="Fish")
plt.plot(data["chronon"], data["nombre de requin"], color="darkblue", label = "Shark")
plt.xlabel('Cycles')
plt.legend()
plt.show()
```

```python
data3 = data.copy()
data3 = data3[50:150]
plt.figure(figsize=(20,10)) 
plt.title("Number of sharks and fish")
plt.plot(data3["chronon"], data3["nombre de poisson"], color="lightblue", label="Fish")
plt.plot(data3["chronon"], data3["nombre de requin"], color="darkblue", label = "Shark")
plt.legend()
plt.xlabel('Cycles')
plt.show()
```

```python

plt.figure(figsize=(20,10)) 
plt.title("Average Lifespan")
plt.plot(data["chronon"], data["average_age_fish"], color="lightblue", label="Fish")
plt.plot(data["chronon"], data["average_age_shark"], color="darkblue", label = "Shark")
plt.legend()
plt.xlabel('Cycles')
plt.show()
```

```python
plt.figure(figsize=(20,10)) 
plt.title("Average Number of Descendants")
plt.scatter(data["chronon"], data["nb_descendants_fish"], color="lightblue", label="Fish")
plt.scatter(data["chronon"], data["nb_descendants_shark"], color="darkblue", label = "Shark")
plt.xlabel('Cycles')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,10)) 
plt.title("Max Number of descendants")
plt.plot(data["chronon"], data["max_nb_descendants_fish"], color="lightblue", label="max nb descendnats Fish")
plt.plot(data["chronon"], data["max_nb_descendants_shark"], color="darkblue", label="max nb descendnats Shark")
plt.xlabel('Cycles')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,10)) 
plt.title('Average Distance Traveled')
plt.plot(data["chronon"], data["average_distance_fish"], color="lightblue", label = "Fish")
plt.plot(data["chronon"], data["average_distance_shark"], color="darkblue", label = "Shark")
plt.xlabel('Cycles')
plt.legend()
plt.show()
```
